```python
import numpy as np # import lib numpy
import cv2  # import lib cv2
import matplotlib.pyplot as plt  #import lib matplotlib
```

```python
img = cv2.imread("patrick.jpg") #membaca gambar dalam folder dengan menggunakan cv2
```

```python
img_height = img.shape[0] # mengekstrak tinggi gambar (0)
img_width = img.shape[1] #mengekstrak lebar gambar (1)
img_channel = img.shape[2] #mengekstrak jumlah channel gambar (2)
```

```python
img_grayscale = np.zeros(img.shape, dtype=np.uint8) # membuat array dengan ukuran yang sama dengan gambar asli dengan tipe data uint8

for y in range(0, img_height):
    for x in range(0, img_width): 
        # Iterasi untuk setiap pixel dalam gambar menggunakan variabel x dan y.
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        #Mengambil nilai intensitas warna merah, hijau, dan biru dari setiap pixel dalam gambar menggunakan indeks pada array.
        gray = (int(red) + int(green) + int(blue)) / 3 #Menghitung nilai intensitas grayscale dari setiap pixel dengan cara menambahkan nilai intensitas warna merah, hijau, dan biru, lalu dibagi tiga.
        img_grayscale[y][x] = (gray, gray, gray) #Menyimpan nilai intensitas grayscale ke dalam array kosong pada indeks yang sesuai untuk setiap pixel.
        
plt.imshow(img_grayscale)
plt.title("Grayscale")
plt.show()
#Menampilkan gambar grayscale yang dihasilkan menggunakan Matplotlib.
```

```python
hg = np.zeros((256))#Membuat array kosong hg dengan panjang 256 menggunakan fungsi np.zeros(). 
```

```python
for x in range(0, 256): #Membuat loop for untuk mengiterasi setiap nilai intensitas piksel yang mungkin dalam gambar grayscale.
    hg[x] = 0 #Mengatur nilai pada elemen array hg dengan indeks yang sesuai dengan nilai iterasi saat ini (yaitu, x) menjadi 0.
```

```python
for y in range(0, img_height):#Membuat dua loop for untuk mengiterasi setiap piksel dalam gambar grayscale.
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0] #Mengambil nilai intensitas piksel grayscale dari gambar pada lokasi (y, x) dan menyimpannya ke variabel gray. 
        hg[gray] += 1 #Menambahkan satu kemunculan intensitas piksel gray ke elemen array hg dengan indeks yang sesuai.
```

```python
# plt.figure(figsize=(20, 6))
# plt.plot(hg, color="black", linewidth=2.0)
# plt.show()

bins = np.linspace(0, 256, 100) #Membuat array bins yang terdiri dari 100 nilai yang sama lebarnya, dengan nilai awal 0 dan nilai akhir 256
plt.hist(hg, bins, color="black", alpha=0.5) #plot histogram
plt.title("Histogram") #plot histogram
plt.show() #tampilkan histogram
```

```python
hgr = np.zeros((256)) #array untuk menyimpan informasi kemunculan intensitas merah dari setiap piksel dalam gambar.
hgg = np.zeros((256)) #array untuk menyimpan informasi kemunculan intensitas hijau dari setiap piksel dalam gambar.
hgb = np.zeros((256)) #array untuk menyimpan informasi kemunculan intensitas biru dari setiap piksel dalam gambar.
hgrgb = np.zeros((768)) #array untuk menyimpan informasi kemunculan kombinasi intensitas saluran warna merah, hijau, dan biru dari setiap piksel dalam gambar.
```

```python
for x in range(0, 256):
    hgr[x] = 0
    hgg[x] = 0
    hgb[x] = 0
#Pada loop pertama, ketiga array hgr, hgg, dan hgb diisi dengan nilai 0 untuk setiap elemennya. Loop ini berjalan sebanyak 256 kali, yang sesuai dengan panjang masing-masing array.
for x in range(0, 768):
#Pada loop kedua, array hgrgb diisi dengan nilai 0 untuk setiap elemennya. Loop ini berjalan sebanyak 768 kali, yang sesuai dengan panjang array hgrgb.
    hgrgb[x] = 0 #masukan nilai 0 pada array hgrgb 
```

```python
for x in range(0, 256): #looping variable x 0-256
    hgr[x] = 0 #masukan 0 pada list hgr
    hgg[x] = 0 #masukan 0 pada list hgg
    hgb[x] = 0  #masukan 0 pada list hgb
    
for x in range(0, 768):
#looping variable x 0-768
    hgrgb[x] = 0 #masukan 0 pada list hgrgb

# th = int(256/64)
temp = [0]  # deklarasi array kosong 'temp'
for y in range(0, img.shape[0]):  # melakukan loop sebanyak tinggi gambar
    for x in range(0, img.shape[1]):  # melakukan loop sebanyak lebar gambar
        # mengambil nilai intensitas warna merah, hijau, dan biru (RGB) pada piksel tertentu
        red = int(img[y][x][0])
        green = int(img[y][x][1])
        blue = int(img[y][x][2])
        # menambahkan offset nilai hijau dan biru agar nilainya tidak sama dengan nilai merah
        green = green + 256
        blue = blue + 512
        # memperbarui nilai histogram RGB untuk setiap nilai intensitas warna (0-255)
        hgrgb[red] += 1
        hgrgb[green] += 1
        hgrgb[blue] += 1


binsrgb = np.linspace(0, 768, 100)#Membuat array bins yang terdiri dari 768 nilai yang sama lebarnya, dengan nilai awal 0 dan nilai akhir 100
plt.hist(hgrgb, binsrgb, color="black", alpha=0.5)#memplot histogram
# plt.plot(hgrgb)
plt.title("Histogram Red Green Blue")#memplot judul
plt.show()#menampilkan gambar 
```

```python
for y in range(0, img_height): #Looping untuk setiap baris gambar (dimulai dari baris ke-0) dengan variabel y.
    for x in range(0, img_width): #Looping untuk setiap piksel pada baris tersebut (dimulai dari piksel ke-0) dengan variabel x.
        red = img[y][x][0] #Mengambil nilai intensitas warna merah pada piksel yang sedang ditinjau dengan indeks warna ke-0 pada piksel tersebut.
        green = img[y][x][1] #Mengambil nilai intensitas warna hijau pada piksel yang sedang ditinjau dengan indeks warna ke-1 pada piksel tersebut.
        blue = img[y][x][2] # Mengambil nilai intensitas warna biru pada piksel yang sedang ditinjau dengan indeks warna ke-2 pada piksel tersebut.
        hgr[red] += 1 #Menambahkan jumlah kemunculan intensitas warna merah ke dalam histogram hgr dengan menggunakan nilai intensitas warna merah sebagai indeksnya.
        hgg[green] += 1 #Menambahkan jumlah kemunculan intensitas warna hijau ke dalam histogram hgg dengan menggunakan nilai intensitas warna hijau sebagai indeksnya.
        hgb[blue] += 1 #Menambahkan jumlah kemunculan intensitas warna biru ke dalam histogram hgb dengan menggunakan nilai intensitas warna biru sebagai indeksnya.

bins = np.linspace(0, 256, 100)
plt.hist(hgr, bins, color="red", alpha=0.5)#plot histogram merah
plt.title("Histogram Red")#plot title
plt.show()#tampikan gambar

plt.hist(hgg, bins, color="green", alpha=0.5)#plot histogram hijau
plt.title("Histogram Green")#plot tittle
plt.show()#tampilkan gambar

plt.hist(hgb, bins, color="blue", alpha=0.5)#plot histogram biru
plt.title("Histogram Blue")#plot tittle
plt.show()#tampilkan gambar
```

```python
#Membuat dua buah array numpy dengan 256 elemen bernilai nol.
hgk = np.zeros((256)) 
c = np.zeros((256))
#nisialisasi elemen-elemen di dalam array hgk dan c menjadi nol.
for x in range(0, 256):
    hgk[x] = 0
    c[x] = 0
#Melakukan iterasi piksel pada citra grayscale, menghitung histogram grayscale dan menyimpannya dalam array hgk.
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgk[gray] += 1
#Menghitung nilai awal untuk array c.                
c[0] = hgk[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgk[x]

hmaxk = c[255]
#Melakukan iterasi untuk menghitung kumulatif histogram dengan menjumlahkan elemen-elemen di dalam array hgk.
for x in range(0, 256):
# Menghitung nilai maksimum kumulatif histogram hmaxk.
#Menghitung nilai kumulatif histogram yang dinormalisasi dengan mengalikan setiap elemen di dalam array c dengan nilai konstanta 190 dan membaginya dengan nilai maksimum hmaxk.

    c[x] = 190 * c[x] / hmaxk

plt.hist(c, bins, color="black", alpha=0.5) #plot histogran
plt.title("Histogram Grayscale Kumulatif") #plot title
plt.show()#tampilkan gambar
```

```python
#Inisialisasi variabel hgh, h, dan c dengan nilai 0 sebanyak 256 elemen.
hgh = np.zeros((256))
h = np.zeros((256))
c = np.zeros((256))
#Mengisi setiap elemen hgh, h, dan c dengan nilai 0 sebanyak 256 elemen.
for x in range(0, 256):
    hgh[x] = 0
    h[x] = 0
    c[x] = 0
#Menghitung histogram gambar grayscale yang diwakili oleh matriks img_grayscale dengan menghitung setiap kemunculan nilai grayscale pada gambar dan menyimpan hasilnya ke variabel hgh
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgh[gray] += 1
#Menghitung fungsi distribusi kumulatif (CDF) dari histogram grayscale yang disimpan di variabel hgh, dan menyimpan hasilnya di variabel h.               
h[0] = hgh[0]
for x in range(1, 256):
     h[x] = h[x-1] + hgh[x]
#Menghitung probabilitas kemunculan nilai grayscale pada gambar dan menyimpan hasilnya di variabel h.
for x in range(0, 256):
     h[x] = h[x] / img_height / img_width
#Melakukan proses histogram equalization pada gambar grayscale dengan melakukan mapping nilai grayscale yang diwakili oleh img_grayscale ke nilai grayscale baru menggunakan variabel h, kemudian menghitung histogram gambar yang sudah terequalisasi dan menyimpan hasilnya di variabel hgh.
for x in range(0, 256):
    hgh[x] = 0
    
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        gray = h[gray] * 255
        hgh[int(gray)] += 1
#Melakukan normalisasi terhadap histogram gambar yang sudah terequalisasi di variabel hgh, kemudian melakukan scaling terhadap histogram tersebut dengan mengalikan setiap nilai histogram dengan 190 dan dibagi dengan nilai maksimum histogram, kemudian hasilnya disimpan di variabel c.
c[0] = hgh[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgh[x]

hmaxk = c[255]

for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk

plt.hist(c, bins, color="black", alpha=0.5) #plot histogram
plt.title("Histogram Grayscale Hequalisasi") #plot title
plt.show()#tampilkan gambar dengan menggunakan lib matplotlib
```
